# **LINEAR REGRESSION ON CAR PURCHASING DATASET**

---



```python
import pandas as pd
df=pd.read_csv("car_purchasing.csv", encoding="latin1")
df1=df.copy()
df1
```

```python
df1.isnull().sum()
```

```python
df1.drop(['customer name', 'customer e-mail', 'country' ,'gender'],axis=1,inplace=True)
```

```python
df1
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(df1.corr(), annot=True,cmap='coolwarm')
plt.show()
```

```python
x=df1.iloc[:,:-1]
y=df1.iloc[:,-1]
```

```python
y=y.values.reshape(-1,1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x=scaler.fit_transform(x)
y=scaler.fit_transform(y)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,shuffle=True,random_state=42)
x_train
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train)
y_pred=lr.predict(x_test)
y_pred
# mean square error and r2 score and plot graph
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred,color='red',marker='.',label='predicted')
plt.grid()
plt.show()
```

```python
#ridge regression and plot graph
from sklearn.linear_model import Ridge
ridge=Ridge()
ridge.fit(x_train,y_train)
y_pred1=ridge.predict(x_test)
print(ridge.score(x_train,y_train))
mse = mean_squared_error(y_test, y_pred1)
r2 = r2_score(y_test, y_pred1)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred1,color='blue',marker='.',label='predicted')

plt.title('y_test vs y_pred')
plt.xlabel('y_test')
plt.ylabel('y_pred')
plt.grid()

plt.show()





```

```python
#lasso regression and plot graph
from sklearn.linear_model import Lasso
lasso=Lasso()
lasso.fit(x_train,y_train)
y_pred2=lasso.predict(x_test)
print(lasso.score(x_train,y_train))
mse = mean_squared_error(y_test, y_pred2)
r2 = r2_score(y_test, y_pred2)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred2,color='green',marker='.',label='predicted')
plt.grid()
plt.show()

```
