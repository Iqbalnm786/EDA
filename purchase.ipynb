# **BUILDING ML MODEL**

---



```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import GaussianNB
```

```python
x_train = train.drop(columns=["Churn"])
y_train = train["Churn"]

x_test=test.copy()
```

```python
cat_cols = x_train.select_dtypes(include="object").columns
num_cols = x_train.select_dtypes(exclude="object").columns
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), num_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols)
    ]
)
```

```python
models = {
    "Logistic Regression": LogisticRegression(
        max_iter=3000, solver="saga", class_weight="balanced", n_jobs=-1
    ),
    "KNN": KNeighborsClassifier(n_neighbors=5),  # keep for comparison
    "Decision Tree": DecisionTreeClassifier(
        max_depth=10, min_samples_leaf=50, class_weight="balanced"
    ),
    "Random Forest": RandomForestClassifier(
        n_estimators=100, max_depth=12, min_samples_leaf=50, n_jobs=-1
    ),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=100),
    "SVM": LinearSVC(class_weight="balanced", max_iter=2000)

}
```

```python
# Convert categorical columns to string
for col in cat_cols:
    x_train[col] = x_train[col].astype(str)
    x_test[col] = x_test[col].astype(str)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.metrics import f1_score, accuracy_score

results = {}

for name, clf in models.items():
    pipe = Pipeline(steps=[
        ("preprocess", preprocessor),
        ("model", clf)
    ])

    pipe.fit(x_train, y_train)
    y_pred = pipe.predict(x_train)   # or validation set

    results[name] = {
        "Accuracy": accuracy_score(y_train, y_pred,),
        "F1 Score": f1_score(y_train, y_pred, average="weighted")
    }

results_df = pd.DataFrame(results).T
print(results_df)
```

```python
best_model_name = results_df["F1 Score"].idxmax()
print("Best Model:" ,best_model_name )
```

```python
best_model = models[best_model_name]
final_pipe = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("model", best_model)
])

final_pipe.fit(x_train, y_train)
y_pred = final_pipe.predict(x_test)
```

```python
test["predicted_churn"] = final_pipe.predict(x_test)
```

```python
test[["predicted_churn"]].to_excel("final.xlsx", index=False)
```
